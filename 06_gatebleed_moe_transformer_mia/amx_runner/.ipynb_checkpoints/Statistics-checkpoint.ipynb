```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv('amx_sniff_data.csv', header=None, names=['label', 'value'])
# df['value'] = df['value'].str.rstrip('%').astype(float)

# Separate data based on label (0 or 1)
values_0 = df[df['label'] == 0]['value']
values_1 = df[df['label'] == 1]['value']

# Plot histograms

plt.figure(figsize=(10, 6))
plt.hist(values_0, bins=30, alpha=0.6, label='Full Model', color='blue', density=True)
plt.hist(values_1, bins=30, alpha=0.6, label='Early Exit', color='orange', density=True)

# Labels and legend
plt.title('Full MIA Attack')
plt.xlabel('% of fast AMX operations during inference')
plt.ylabel('Frequency')
plt.legend()
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np
print(f"Expert 1 mean: {np.mean(values_0)}")
print(f"Expert 2 mean: {np.mean(values_1)}")
print("")
print(f"Expert 1 std: {np.std(values_0)}")
print(f"Expert 2 std: {np.std(values_1)}")


print("")
# Calculate mean and standard error
mean = np.mean(values_0)
sem = stats.sem(values_0)  # standard error of the mean
# Compute the 95% confidence interval
confidence = 0.95
h = sem * stats.t.ppf((1 + confidence) / 2., len(values_0) - 1)
ci_lower = mean - h
ci_upper = mean + h
print(f"Expert 1 95% CI: [{ci_lower}, {ci_upper}]")

# Calculate mean and standard error
mean = np.mean(values_1)
sem = stats.sem(values_1)  # standard error of the mean
# Compute the 95% confidence interval
confidence = 0.95
h = sem * stats.t.ppf((1 + confidence) / 2., len(values_1) - 1)
ci_lower = mean - h
ci_upper = mean + h
print(f"Expert 2 95% CI: [{ci_lower}, {ci_upper}]")

midpoint = (np.mean(values_0) + np.mean(values_1)) / 2
fp = np.sum(values_0 < midpoint) / len(values_0)
fn = np.sum(values_1 > midpoint) / len(values_1)
print(f"False positive: {fp*100}%")
print(f"False negative: {fn*100}%")
```
